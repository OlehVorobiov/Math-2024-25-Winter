```python
import numpy as np
import matplotlib.pyplot as plt

def estimate_pi(num_points, plot=True):
    # Generate random x and y coordinates in range [-1, 1]
    x = np.random.uniform(-1, 1, num_points)
    y = np.random.uniform(-1, 1, num_points)

    # Compute distance from origin and count how many are inside the circle
    distances_squared = x**2 + y**2
    inside_circle = distances_squared <= 1
    pi_estimate = 4 * np.sum(inside_circle) / num_points

    # Optional plot
    if plot:
        fig, ax = plt.subplots(figsize=(6, 6))
        ax.set_aspect('equal')
        ax.scatter(x[inside_circle], y[inside_circle], color='blue', s=1, label='Inside Circle')
        ax.scatter(x[~inside_circle], y[~inside_circle], color='red', s=1, label='Outside Circle')
        ax.set_title(f"Monte Carlo Estimation of  with {num_points} Points\nEstimated   {pi_estimate:.6f}")
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        ax.legend()
        plt.grid(True)
        plt.show()

    return pi_estimate

# Example usage:
estimate_pi(num_points=10000)
```
