```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For prettier plots
sns.set(style="whitegrid")

# Settings
population_size = 100_000
sample_sizes = [5, 10, 30, 50]
num_samples = 1000

# Distribution generators
def generate_population(dist_type):
    if dist_type == "uniform":
        return np.random.uniform(0, 1, population_size)
    elif dist_type == "exponential":
        return np.random.exponential(scale=1.0, size=population_size)
    elif dist_type == "binomial":
        return np.random.binomial(n=10, p=0.5, size=population_size)
    else:
        raise ValueError("Unknown distribution type")

# Run simulations and plot
def run_simulation(distribution_name):
    print(f"\nProcessing distribution: {distribution_name.upper()}")
    population = generate_population(distribution_name)

    for n in sample_sizes:
        sample_means = []

        for _ in range(num_samples):
            sample = np.random.choice(population, size=n, replace=False)
            sample_means.append(np.mean(sample))

        # Plotting
        plt.figure(figsize=(8, 5))
        sns.histplot(sample_means, kde=True, bins=30, color="skyblue")
        plt.title(f"{distribution_name.capitalize()} Distribution - Sample Size n={n}")
        plt.xlabel("Sample Mean")
        plt.ylabel("Frequency")
        plt.axvline(np.mean(population), color='red', linestyle='--', label='Population Mean')
        plt.legend()
        plt.tight_layout()
        plt.show()

# Run for all distributions
distributions = ["uniform", "exponential", "binomial"]
for dist in distributions:
    run_simulation(dist)
```
