```python
import sympy as sp
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from sympy import symbols, Eq, expand, solve
import numpy as np

class EnhancedEq(Eq):
    def __add__(self, other):
        return EnhancedEq(self.lhs + other.lhs, self.rhs + other.rhs)

    def __sub__(self, other):
        return EnhancedEq(self.lhs - other.lhs, self.rhs - other.rhs)

    def __mul__(self, scalar):
        return EnhancedEq(scalar * self.lhs, scalar * self.rhs)

    def __rmul__(self, scalar):
        return self.__mul__(scalar)

    def simplify(self):
        return EnhancedEq(expand(self.lhs), expand(self.rhs))

    def substitute(self, *args, **kwargs):
        return EnhancedEq(self.lhs.subs(*args, **kwargs), self.rhs.subs(*args, **kwargs))

    def solve_for(self, symbol):
        return solve(self, symbol)

#Find the equation of a circle with center at point  A(1,2)  and radius  r=3


A = sp.Point(1, 2)
r = 3

x, y = symbols('x y')
eq = EnhancedEq((x-A.x)**2 + (y-A.y)**2, r**2)

print(eq)
print("The equartion is (x -", A.x, ")^2 + (y -", A.y, ")^2 = ", r**2)



circle = Circle(A, r, facecolor='none', edgecolor='blue')

# Create a figure and axes
fig, ax = plt.subplots()

# Add the circle patch to the axes
ax.add_patch(circle)

# Set axis limits (to ensure the full circle is visible)
ax.set_xlim([-3, 6])
ax.set_ylim([-3, 6])

# Set aspect ratio to 'equal' for a circular shape
ax.set_aspect('equal')

# Display the plot
plt.title('Circle')
plt.xlabel('x')
plt.ylabel('y')

min_val = min(-3, 6)
max_val = max(-3, 6)
common_range = np.arange(min_val, max_val + 1)

# Set the grid lines using the common range
ax.set_xticks(common_range)
ax.set_yticks(common_range)
ax.grid(True)

plt.show()
```

```python
#Find the equation of a parabola intersecting the  Ox  axis at points  x=2 ,  x=4 , and passing through point  y(3)=1 .



#x1 = 2 =>
A = sp.Point(2, 0)
#x2 = 4 =>
B = sp.Point(4, 0)

#y(3) = 1 =>
C = sp.Point(3, 1)

a, x, y = symbols('a x y')
#Establishing equation of parabola through given points
eq = EnhancedEq(a*(x - A.x)*(x - B.x), y)  #The formula of the parabola
print(eq)

#Searching for a
eq1 = eq.substitute(x, C.x,)
eq1 = eq1.substitute(y, C.y,)
print(eq1)
a_sol = eq1.solve_for(a)[0]
print(a_sol)

eq = eq.substitute(a, a_sol)
print(eq)

eq = eq.simplify()
print(eq)

print("The equartion is y = -x^2 + 6x - 8")


def parabola(x):
    return -x**2 + 6*x - 8

# Create x-values
x = np.linspace(0, 6, 100)  # Adjust range as needed

# Calculate corresponding y-values
y = parabola(x)


# Plot the parabola
plt.plot(x, y)
plt.title('Parabola: y = -x^2 + 6x - 8')
plt.xlabel('x')
plt.ylabel('y')


plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

plt.xlim(min(0, 0), max(6, 6))
plt.ylim(min(-5, -5), max(2, 2))

plt.show()
```

```python
#Find the center of the ellipse with the equation  x2+4y24x16y+16=0 .
from matplotlib.patches import Ellipse

x, y = symbols('x y')

#The initial equation
eq = EnhancedEq(x**2 + 4*y**2 - 4*x - 16*y + 16, 0)
print(eq)

#Simplifying the equation
eq = EnhancedEq((x-2)**2 + 4*(y-2)**2, 4)
print(eq)

O = sp.Point(2, 2)
print("The center is",O)

ellipse = Ellipse(O, 2*2, 2*1, facecolor='none', edgecolor='blue')  # width=2*a, height=2*b

# Create a figure and axes
fig, ax = plt.subplots()

# Add the circle patch to the axes
ax.add_patch(ellipse)

# Set axis limits (to ensure the full circle is visible)
ax.set_xlim([-1, 4])
ax.set_ylim([-1, 4])

# Set aspect ratio to 'equal' for a circular shape
ax.set_aspect('equal')

# Display the plot
plt.title('Circle')
plt.xlabel('x')
plt.ylabel('y')

min_val = min(-1, 5)
max_val = max(-1, 5)
common_range = np.arange(min_val, max_val + 1)

# Set the grid lines using the common range
ax.set_xticks(common_range)
ax.set_yticks(common_range)
ax.grid(True)

ax.scatter(O.x, O.y, color='red', marker='o', s=50, label = "center")
plt.legend()
plt.show()
```

```python
#Find the slope ( m>0 ) of the line  y=mx5  that is tangent to the circle with the equation  x2+y2=1 .
import math
m, x, y = symbols('m x y')

#The initial equation
Circle_ = EnhancedEq(x**2 + y**2 - 1, 0)
Line = EnhancedEq(m*x - 5, y)
print("Circle:", Circle_)
print("Line", Line)

#Solving
LineY = Line.solve_for(y)[0]
Circle_ = Circle_.substitute(y, LineY)
print(Circle_)
Circle_ = Circle_.simplify()
print(Circle_)

D = (-10*m)**2 - 4 * 24*(m**2+1)
print("D:", D)

eq = EnhancedEq(D, 0)
m_sol = eq.solve_for(m)[0]
print("The slope 1 is ",-m_sol)
print("The slope 2 is ",m_sol)
print("The slope is greater than 0 => m =",-m_sol)



circle = Circle([0, 0], 1, facecolor='none', edgecolor='blue')

# Create a figure and axes
fig, ax = plt.subplots()

# Add the circle patch to the axes
ax.add_patch(circle)

# Set axis limits (to ensure the full circle is visible)
ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])

# Set aspect ratio to 'equal' for a circular shape
ax.set_aspect('equal')

# Display the plot
plt.title('Circle x Line')
plt.xlabel('x')
plt.ylabel('y')

min_val = min(-4, 4)
max_val = max(-4, 4)
common_range = np.arange(min_val, max_val + 1)



def line_equation(x, s):
    return s*2 * math.sqrt(6) * x - 5

# Create x-values for the line
x_line = np.linspace(-2, 8, 100)  # Adjust range as needed

# Calculate corresponding y-values using the line equation
y_line = line_equation(x_line, 1)
# Plot the line
ax.plot(x_line, y_line, color='green', linewidth=2)

# Set the grid lines using the common range
ax.set_xticks(common_range)
ax.set_yticks(common_range)
ax.grid(True)

plt.show()
```

```python
#Find the intersection points of the hyperbola  x2y2=1  with the ellipse's line  x2+4y2=6
#The initial equations
x, y = symbols('x y')

hyperbola = EnhancedEq(x**2 - y**2, 1)
ellipse = EnhancedEq(x**2 + 4*y**2, 6)
print("Hyperbola:", hyperbola)
print("Ellipse:", ellipse)

#Solving
ySq = hyperbola.solve_for(y**2)[0]
print(ySq)

eq = ellipse.substitute(y**2, ySq)
print(eq)

x_sol1 = eq.solve_for(x)[0]
print(x_sol1)
x_sol2 = -eq.solve_for(x)[0]
print(x_sol2)

eq1 = hyperbola.substitute(x, x_sol1)
y_sol1 = eq1.solve_for(y)[0]
print(y_sol1)
eq2 = hyperbola.substitute(x, x_sol2)
y_sol2 = -eq2.solve_for(y)[0]
print(y_sol2)

print("The intersection points are (",x_sol1,",",y_sol1,"), (",x_sol2,",",y_sol2,"), (",x_sol1,",",y_sol2,") and (",x_sol2,",",y_sol1,")")


# Define the range for x and y
x = np.linspace(-5, 5, 100)  # Adjust the range as needed
y = np.linspace(-5, 5, 100)  # Adjust the range as needed
X, Y = np.meshgrid(x, y)

# Define the equation of the ellipse (x^2 + 4y^2 = 6)
equation = X**2 + 4*Y**2 - 6  # Equation in the form f(x, y) = 0

# Plot the ellipse using contour
plt.contour(X, Y, equation, [0])  # [0] represents the level curve for equation = 0

x = np.linspace(-5, 5, 100)  # Adjust the range as needed
y = np.linspace(-5, 5, 100)  # Adjust the range as needed
X, Y = np.meshgrid(x, y)

# Define the equation of the hyperbola (x^2 - y^2 = 1)
equation = X**2 - Y**2 - 1  # Equation in the form f(x, y) = 0

# Plot the hyperbola using contour
plt.contour(X, Y, equation, [0])  # [0] represents the level curve for equation = 0


plt.scatter(x_sol1, y_sol1, color='red', marker='o', s=50)
plt.scatter(x_sol2, y_sol2, color='red', marker='o', s=50)
plt.scatter(x_sol2, y_sol1, color='red', marker='o', s=50)
plt.scatter(x_sol1, y_sol2, color='red', marker='o', s=50)

# Set aspect ratio to 'equal' for a proper ellipse shape
plt.gca().set_aspect('equal')

# Add labels and grid
plt.title('Ellipse x Hyperbola')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

```python
#For the given hyperbola  x2y2=1 , find the distance between its branches.
import math
x, y = symbols('x y')

hyperbola = EnhancedEq(x**2 - y**2, 1)
print("Hyperbola:", hyperbola)

# Define symbols
x, y = symbols('x y')

# Equation of the hyperbola
hyperbola = sp.Eq(x**2 - y**2, 1)

# Solve for x (branches of the hyperbola)
x_solutions = sp.solve(hyperbola, x)

# The distance between the branches is 2*sqrt(y^2 + 1)
distance_expr = 2 * sp.sqrt(y**2 + 1)

# Calculate the minimum distance by substituting y = 0
min_distance = distance_expr.subs(y, 0)

# Print the result
min_distance

print("The minimum distance between the branches is:", min_distance)


# Generate y values
y_values = np.linspace(-3, 3, 400)

# Calculate x values for the branches
x_positive = np.sqrt(y_values**2 + 1)
x_negative = -np.sqrt(y_values**2 + 1)

# Calculate the distance between the branches (horizontal distance)
distance_values = 2 * np.sqrt(y_values**2 + 1)

# Plot the branches and the distance for comparison
plt.figure(figsize=(10, 6))

# Plot the two branches
plt.plot(x_positive, y_values, label="Branch 1 (x = +sqrt(y^2 + 1))", color='blue')
plt.plot(x_negative, y_values, label="Branch 2 (x = -sqrt(y^2 + 1))", color='blue')

# Highlight the minimum distance (at y = 0)
plt.axhline(0, color='red', linestyle='--', label='y = 0 (minimum distance)')

# Plot the distance between the branches as a function of y
plt.plot(y_values, distance_values, label='Distance between branches', color='green', linestyle='--')

# Labeling
plt.title('Minimum Distance and Branches of the Hyperbola x^2 - y^2 = 1')
plt.xlabel('y')
plt.ylabel('Distance')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
