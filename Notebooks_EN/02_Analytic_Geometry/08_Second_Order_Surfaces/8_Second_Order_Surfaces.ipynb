* Write the equation of a sphere with center at point $P=(1,2,3)$ and radius $r=3$.
* Do the spheres with equations $x^2 + y^2 + z^2 = 1$ and $x^2 + y^2 + z^2 = 2$ have any common points?
* What curve in space is formed by the intersection of the sphere $x^2 + y^2 + z^2 = 1$ with the sphere $(x-1)^2 + y^2 + z^2 = 1$? Find the equation of this curve.
* Write the equation of the tangent plane to the paraboloid $z=(x-1)^2+y^2+1$ at point $P=(1,0,1)$.

```python
#Write the equation of a sphere with center at point  P=(1,2,3)  and radius  r=3 .

import sympy as sp
import numpy as np

x, y, z = sp.symbols('x y z')

P = np.array([1,2,3])
r = 3

Eq = sp.Eq((x-P[0])**2 + (y-P[1])**2 + (z-P[2])**2, r**2)
print(Eq)
print("The equation of the sphere is (x-1)^2+(y-2)^2+(z-3)^2=9")
```

```python
#Do the spheres with equations  x2+y2+z2=1  and  x2+y2+z2=2  have any common points?

x, y, z = sp.symbols('x y z')

Eq1 = sp.Eq(x**2 + y**2 + z**2, 1)
Eq2 = sp.Eq(x**2 + y**2 + z**2, 2)

print(Eq1)
print(Eq2)

print("The spheres do not have any common points, since they both have the same center and differents radii")
```

```python
#What curve in space is formed by the intersection of the sphere  x2+y2+z2=1
#with the sphere  (x1)2+y2+z2=1 ? Find the equation of this curve.

from sympy import symbols, Eq, expand, solve

class EnhancedEq(Eq):
    def __add__(self, other):
        return EnhancedEq(self.lhs + other.lhs, self.rhs + other.rhs)

    def __sub__(self, other):
        return EnhancedEq(self.lhs - other.lhs, self.rhs - other.rhs)

    def __mul__(self, scalar):
        return EnhancedEq(scalar * self.lhs, scalar * self.rhs)

    def __rmul__(self, scalar):
        return self.__mul__(scalar)

    def simplify(self):
        return EnhancedEq(expand(self.lhs), expand(self.rhs))

    def substitute(self, *args, **kwargs):
        return EnhancedEq(self.lhs.subs(*args, **kwargs), self.rhs.subs(*args, **kwargs))

    def solve_for(self, symbol):
        return solve(self, symbol)

x, y, z = sp.symbols('x y z')

Eq1 = EnhancedEq(x**2 + y**2 + z**2, 1)
Eq2 = EnhancedEq((x-1)**2 + y**2 + z**2, 1)

Eq2 = EnhancedEq(-2*x + 1, 0) #Substituting part of the equation with the first equation
x_sol = Eq2.solve_for(x)[0]

Eq1 = Eq1.substitute(x, x_sol) #The equation
r2 = Eq1.solve_for(y**2 + z**2)[0]


print("The formed curve is a circle")
print("The equation of the curve is y^2 + z^2 =", r2)




```
