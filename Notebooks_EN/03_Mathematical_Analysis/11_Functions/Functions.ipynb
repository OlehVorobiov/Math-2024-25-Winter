```python
import sympy as sp
import math

f = 2**2
g = math.sqrt(2)
h = 1/2
j = math.sin(2)

print("f(2) =",f)
print("g(2) =", round(g,2))
print("h(2) =", h)
print("j(2) =", round(j, 2))
```

```python
#Let  f(x)=3x−1  and  g(x)=x−−√ . Find:
#f(g(x))
#g(f(x))
#f(f(x))
#g(g(x))
#and visualize functions in a single Geogebra notebook.

x = sp.symbols("x")

f = 3*x-1
g = sp.sqrt(x)

print("f(g(x)) =", f.subs(x, g))
print("g(f(x)) =", g.subs(x, f))
print("f(f(x)) =", f.subs(x, f))
print("g(g(x)) =", g.subs(x, g))
```

```python
#Let  f(x)=ex  and  g(x)=ln(x) . Check:  f(g(x))  and  g(f(x)) . What do you notice?
x = sp.symbols("x")
f = math.e*x
g = sp.log(x)

print("f(g(x)) =", f.subs(x, g))
print("g(f(x)) =", g.subs(x, f))
```
