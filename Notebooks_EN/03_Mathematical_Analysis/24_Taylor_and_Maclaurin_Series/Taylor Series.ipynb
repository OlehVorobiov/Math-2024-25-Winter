1. Find the Taylor series and visualize obtained functions in Geogebra:
   - $f(x) = \cos(x)$ around $x = 0$ up to the 4th degree.

To solve this, we can use the formula of Taylor series:

$f(x) = f(a) + \frac{f'(a)}{1!}(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \frac{f^{(3)}(a)}{3!}(x - a)^3 + \cdots
$

$a = 0$

   $f(x) = \cos(0) - sin(0)(x - 0) - \frac{cos(0)(x-0)^2}{2!} + \frac{sin(0)(x-0)^3}{3!} + \frac{cos(0)(x-0)^4}{4!}$

  $f(x) = 1 - \frac{x^2}{2} + \frac{x^4}{24}$

- $h(x) = 1/(1-x)$ around $x = 0$ up to the 4rd degree.

$a = 0$

$h(x) = \frac{1}{1-a} + \frac{x}{(1-a)^2} + \frac{x^2}{(1-a)^3} + \frac{x^3}{(1-a)^4} + \frac{x^4}{(1-a)^5}$

$h(x) = 1 + x + x^2 + x^3 + x^4$

- $g(x) = \sin(x)$ around $x = \pi$ up to the 4rd degree.

$g'(x) = cos(x)$   $ => g'(\pi) = -1$

$g''(x) = -sin(x)$ $ => g''(\pi) = 0$

$g'''(x) = -cos(x)$ $ => g'''(\pi) = 1$

$g''''(x) = sin(x)$ $ => g''''(\pi) = 0$

$g(x) = -x+\pi-\frac{-(x-\pi)^3}{6}$

Find a tangent line $y = f'(x_0) (x-x_0) + f(x_0)$ to the function $f(x) = e^{\sin(x)}$ at $x_0 = \pi$. Hints for Geogebra visualization: define f(x), include slider s, define y = f'(s) (x-s) + f(s), and include point P(s, f(s)).

$\text{The equation of tangent line is essentially Taylor Series equation up to the 1st degree}$

$f(x) = f(a) + \frac{f'(a)}{1!}(x - a)$

$\text{To solve it, we need to find the unknown parts of the equation}$

$f(x) = e^{\sin(x)} => f(x_0) = e^{\sin(\pi)} = e^0 = 1$

$f'(x) = e^{\sin(x)}*cos(x) => f'(x) = e^{\sin(\pi)}*cos(\pi) = -1$

$y = 1-1(x-\pi)$

$y = -x+\pi+1$
