5. Calculate the length of the sine curve over the same interval using:

$$L = \int_a^b \sqrt{1 + (f'(x))^2} \, dx= \int_0^\pi \sqrt{1 + \cos^2(x)} \, dx
$$

$f(x) = \sin(x)$

$I = [0,\pi]$

$f'(x) = \cos(x)$

$L = \int^\pi_0\sqrt{1+\cos^2(x)} dx$

The term $\sqrt{1+\cos^2(x)}$ oscillates because $\cos^2(x)$ varies between 0 and 1:

1. At $x = 0: cos^2(0) = 1$, so $\sqrt{1+cos^2(x)} = \sqrt{2}$

2. At $x = \frac{\pi}{2}: cos^2(\frac{\pi}{2}) = 1$, so $\sqrt{1+cos^2(x)} = 1$

3. At $x = \pi: cos^2(\pi) = 1$, so $\sqrt{1+cos^2(x)} = \sqrt{2}$

$\sqrt{1+\cos^2(x)}$ oscillates between 1 and $\sqrt{2}$

Evaluating numerically, since the integral does not have an elementary antiderivative:

$L = \int^\pi_0\sqrt{1+\cos^2(x)} dx ≈ 3.82$

6. Find the distance of the moving particle between time $t=0$ and $t=2$ for the following position function: $x(t) = 3t^2 - 6t + 1$.

In order to find the distance, we need to find the velocity of the function:

$v(d) = x'(t) = 6t - 6$

Check if the particle changes direction:

$6t - 6 = 0$

The direction changes at $t = 1$

The total distance traveled is the integral of velocity:

   $\int_a^b f(t) \, dt = F(t2)-F(t1)$
   
**Total Distance Traveled** $= -\int_0^1(6t - 6)dt + \int_1^2(6t - 6)dt= -(3-6) + 3*4 - 6*2 - (3-6) = 3 + 3 = 6$
