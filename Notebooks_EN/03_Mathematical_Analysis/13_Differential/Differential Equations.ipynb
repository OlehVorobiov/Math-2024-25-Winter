**Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}$:**

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

Solving the wave equation:

\[
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
\]

Assume a solution of the form:

$
\psi(t, x) = A \cos(\omega t + kx).
$

Compute the second derivative with respect to \( t \)

$
\frac{\partial \psi}{\partial t} = -A \omega \sin(\omega t + kx).
$

$
\frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx).
$

$
\frac{\partial^2 \psi}{\partial t^2} = -\omega^2 \psi.
$

Compute the second derivative with respect to \( x \)

$
\frac{\partial \psi}{\partial x} = -A k \sin(\omega t + kx).
$

$
\frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx).
$

$
\frac{\partial^2 \psi}{\partial x^2} = -k^2 \psi.
$

Substitute into the wave equation

$
-\omega^2 \psi + v^2 k^2 \psi = 0.
$

Factor out \( \psi \):

$
(\omega^2 - v^2 k^2) \psi = 0.
$

Since \( \psi \neq 0 \), we require:

$
\omega^2 = v^2 k^2.
$

Verify the given relationship for \( v \)

$
v = \frac{\omega}{k}.
$

Squaring both sides:

$
v^2 = \frac{\omega^2}{k^2}.
$

Multiplying by \( k^2 \):

$
v^2 k^2 = \omega^2.
$

Since this holds, we conclude that:

$
\psi(t, x) = A \cos(\omega t + kx)
$

is a solution of the wave equation.
